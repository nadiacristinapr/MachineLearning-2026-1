```python
import pandas as pd

# Supondo que os dataframes df2010, df2011, ..., df2023 já estejam carregados

# Lista para armazenar os DataFrames com a coluna 'ano' adicionada
dfs = []

# Iterando de 2010 a 2023 para adicionar a coluna 'ano' e armazenar os DataFrames
for ano in range(2018, 2025):
    # Cria uma cópia do DataFrame atual e adiciona a coluna 'ano'
    df = globals()[f"df{ano}"].copy()  # Acessa o DataFrame usando globals()
    df['ano'] = ano  # Adiciona a coluna 'ano' com o valor correspondente
    dfs.append(df)  # Adiciona o DataFrame modificado à lista

# Concatenando todos os DataFrames na lista 'dfs' em um único DataFrame
df_completo = pd.concat(dfs, ignore_index=True)

# Visualizando o resultado
df_completo
```
