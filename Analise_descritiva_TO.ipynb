##Proporção dos preditores e desfecho

```python
for coluna in lista_variaveis:
    print(f"\n==== Variável: {coluna} ====")

    # Frequência absoluta
    freq_abs = df[coluna].value_counts(dropna=False)

    # Frequência relativa (em %)
    freq_rel = df[coluna].value_counts(normalize=True, dropna=False) * 100

    # Juntar em um único DataFrame para visualização
    tabela = pd.DataFrame({
        'Frequência Absoluta': freq_abs,
        'Frequência Relativa (%)': freq_rel.round(2)
    })

    print(tabela)
```

```python
# Listas
lista_variavel_desfecho = ['cura']

lista_preditores = ['ano_grupo','pop_grupo','front', 'cs_sexo', 'cs_raca', 'raiox_tora', 'agravalcoo',
       'agravdiabe', 'agravdoenc', 'bacilosc_e', 'cultura_es', 'bacilosc_2',
       'bacilosc_4', 'tratsup_at', 'pop_liber', 'pop_rua', 'pop_imig',
       'benef_gov', 'agravdroga', 'agravtabac',
       'faixa_etaria2', 'contato_maior5', 'escola_faixa',
       'tempo_diag_trat_14']

# Dicionário para salvar os resultados
resultados = {}

desfecho = lista_variavel_desfecho[0]

for var in lista_preditores:

    # Versão temporária sem NA (convertendo para object)
    temp_var = df[var].astype('object').where(df[var].notna(), 'NA')

    # Contagem
    tabela = (
        pd.DataFrame({var: temp_var, desfecho: df[desfecho]})
        .groupby([var, desfecho], dropna=False)
        .size()
        .rename("Frequência Absoluta")
        .reset_index()
    )

    # Frequência relativa
    tabela["Frequência Relativa (%)"] = (
        tabela["Frequência Absoluta"]
        / tabela.groupby(var)["Frequência Absoluta"].transform("sum")
        * 100
    ).round(2)

    resultados[var] = tabela

# Exibir
for var, tabela in resultados.items():
    print(f"\n==== Variável: {var} ====")
    print(tabela)
```
