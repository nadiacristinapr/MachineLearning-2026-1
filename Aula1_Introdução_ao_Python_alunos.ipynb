# 3) Criando Variáveis

Em Python, uma das primeiras coisas que aprendemos é **atribuir valores a um nome** → ou seja, guardar uma informação → variável.

→ É como criar uma pequena "etiqueta" que aponta para um valor.

- **Como fazemos isso?** → usamos o sinal de igual (`=`) para atribuir valores a uma variável:

```python
idade = 35
```

→ Aqui, a palavra `idade` passa a representar o número **35**. Dizemos que “atribuímos o valor 35 à variável `idade`”.

- **Outro exemplo:**

```python
nome = "Maria"
```

→ Agora, `nome` guarda o texto `"Maria"`.

---
Em Python, usamos **aspas** para definir **textos (strings)**.

Você pode usar **aspas simples** `'texto'` ou **aspas duplas** `"texto"` — ambas funcionam!


Atenção: Não misture os dois tipos no mesmo texto:

---

**Regras para nomear variáveis:**
- Começar com **letra**
- **Sem espaços** → use underline: `idade_paciente`
- **Não usar símbolos especiais** como `@`, `#`, `%`, etc.
- Python diferencia **maiúsculas de minúsculas** → `Idade` é diferente de `idade`


# 4) Principais Tipos de Variáveis

| TIPO 🏷️ | DESCRIÇÃO 📝                                                               | EXEMPLO 💡                                                  |
| :------ | :------------------------------------------------------------------------- | :------------------------------------------------------- |
| `int`   | **Números inteiros** (sem decimais). Para contagens, IDs, quantidades.     | `idade = 35`, `num_leitos = 50`, `id_paciente = 12345`   |
| `float` | **Números com decimais**. Essencial para medidas, taxas, médias.           | `peso_kg = 72.5`, `temperatura = 38.2`, `taxa_incidencia = 0.05` |
| `str`   | **Textos**. Nomes, diagnósticos, descrições.                               | `"Maria Silva"`, `'Dengue'`, `"Cardiologia"`            |
| `bool`  | **Valores lógicos**: **`True`** (verdadeiro) ou **`False`** (falso). Para estados ou condições. | `tem_febre = True`, `vacinado = False`                   |

<h3> Tuplas (tuple) </h3>

Parecidas com listas, mas não podem ser modificadas (são imutáveis).

<h3> Conjuntos (set) </h3>

Ótimos para eliminar duplicatas ou verificar interseções entre conjuntos.

| Tipo    | Estrutura  | Características principais           |

| `tuple` | Tupla      | Ordenada, **imutável**               |

# 7) Operadores Lógicos

Em vez de usar **funções ou métodos**, Python usa **operadores** para comparar valores e verificar condições.

Esses operadores são **símbolos** colocados entre dois valores — e retornam um resultado lógico: `True` (verdadeiro) ou `False` (falso).

---

<h3>Principais operadores lógicos e relacionais</h3>

A tabela abaixo mostra o que cada operador faz e o resultado usando esses valores:

```python
a = 1  
b = 2

````
| Operador 🔣 | Significado 📝                            | Exemplo 💡               | Resultado 🎯 |
|-------------|--------------------------------------------|---------------------------|--------------|
| `==`        | Igual a                                    | `a == b`                  | `False`      |
| `!=`        | Diferente de                               | `a != b`                  | `True`       |
| `>`         | Maior que                                  | `a > b`                   | `False`      |
| `>=`        | Maior ou igual a                           | `a >= b`                  | `False`      |
| `<`         | Menor que                                  | `a < b`                   | `True`       |
| `<=`        | Menor ou igual a                           | `a <= b`                  | `True`       |
| `and`       | **E lógico** (ambas condições verdadeiras) | `a > 0 and b > 0`         | `True`       |
| `or`        | **Ou lógico** (uma condição verdadeira)    | `a > 0 or b < 0`          | `True`       |
| `not`       | **Negação** (inverte o valor lógico)       | `not (a == b)`            | `True`       |


#8) Condicionais

Nem sempre queremos que o código execute tudo de uma vez. Muitas vezes, queremos que determinadas ações aconteçam só se certas condições forem verdadeiras.

Para isso, usamos as estruturas condicionais, também chamadas de "if" (se).

---

Estrutura básica:

```python
if condição:
    # instruções se a condição for verdadeira
```

Exemplo:

---

Adicionando alternativas: else

O bloco else é usado quando a condição não for satisfeita:

---
Verificando múltiplas condições: elif (senão se)

Você pode encadear condições com o elif:

# 9) Funções

À medida que nossos códigos crescem, é comum precisarmos repetir cálculos ou blocos de instruções. Em vez de escrever tudo de novo, podemos criar uma função.

Uma função é como uma “caixa de tarefas”: você dá uma entrada (informações), ela realiza um processo, e devolve uma saída (resultado).

Ela nos ajuda a:
- Organizar melhor o código;
- Evitar repetições;
- Facilitar a reutilização do que já foi feito.

---

<h3> Como criar uma função?

Usamos a palavra-chave def, escolhemos um nome para a função e definimos os parâmetros (informações que ela precisa receber):

```python
def nome_da_funcao(parametro1, parametro2):
    # instruções
    return resultado
```

---
<h3> Exemplo prático


# 10) Bibliotecas

As bibliotecas (em inglês, _modules_ ou _packages_) são conjuntos de ferramentas e funções com utilidades parecidas que facilitam a nossa programação.


As bibliotecas são desenvolvidas por usuários da comunidade python e existem inúmeras bibliotecas disponíveis para utilizarmos

---

<h3>Como utilizar as bibliotecas?

- Para utilizar uma biblioteca precisamos instalá-la e importá-la.  
- Algumas bibliotecas já vem instaladas no **Google Colab**, enquanto outras precisam ser instaladas.
- Para instalar as bibliotecas, utilizamos o comando `!pip install [nome_da_biblioteca]` em uma caixa de texto (também chamada de **chunk**)

<h3> Importação

- Para importar uma biblioteca, utilizamos o comando `import [nome_da_biblioteca]`.  
- Também podemos dar um apelido para a biblioteca a fim de encurtar o seu nome e facilitar a utilização no código. Para isso, utilizamos o comando `import [nome_da_biblioteca] as [apelido_da_biblioteca].`

---

<h3> Exemplo de Bibliotecas

- Para trabalhar com tabelas e dados:  
  - **Pandas:** https://pandas.pydata.org/docs/user_guide/index.html  
  - **NumPy:** https://numpy.org/doc/stable/user/index.html#user
- Para gerar gráficos e visualizações:  
  - **Seaborn:** https://seaborn.pydata.org/examples/index.html  
  - **Matplotlib.pyplot:** https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.html
  - **Plotly:** https://plotly.com/python/
- Para trabalhar com machine learning e modelagem estatística:  
  - **Scikit-learn (sklearn):** https://scikit-learn.org/stable/user_guide.html

Exemplo de instalação


Exemplo de importação


---
# **Materiais Recomendados**

📚 Livros (em português):

* Python para Análise de Dados – Wes McKinney. 2ª edição. O’Reilly Media, 2018.
Focado em manipulação de dados com pandas, excelente para análises em saúde.

* Python para Ciência de Dados – Jake VanderPlas. 1ª edição. O’Reilly Media, 2017.
Introduz todo o ecossistema científico do Python, com NumPy, Matplotlib, Scikit-learn e mais.

* Mãos à Obra: Aprendizado de Máquina com Scikit-Learn e TensorFlow – Aurélien Géron. 2ª edição (em português). O’Reilly Media, 2020.
Um guia prático sobre machine learning, ideal para quem quer dar os primeiros passos com algoritmos reais.


🎥 Canais no YouTube:

*  Laboratório de Big Data e Análise Preditiva em Saúde - LABDAPS/USP: https://www.youtube.com/c/Laborat%C3%B3rioLabdapsUSP Vídeos sobre inteligência artificial, saúde pública e dados aplicados à gestão.

* Programação Dinâmica: https://www.youtube.com/@pgdinamicaCanal focado em ciência de dados e Python com linguagem acessível.

* Attekita Dev: https://www.youtube.com/@attekitadevConteúdo sobre Python, lógica de programação e projetos simples para iniciantes.

* Curso em Vídeo – Curso Completo de Python (Gustavo Guanabara): https://www.youtube.com/watch?v=S9uPNppGsGo&list=PLHz_AreHm4dlKP6QQCekuIPky1CiwmdI6Curso gratuito completo, didático e em português. Muito indicado para iniciantes.

📖 Para consulta e aprofundamento:

* Documentação oficial do Python: https://docs.python.org/3/
