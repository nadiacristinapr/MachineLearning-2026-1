```python
#TRABALHAR COM AS DATAS

# Selecionando as colunas específicas
df_selecionado = df[['DT_NOTIFIC', 'DT_DIAG', 'DT_INIC_TR', 'DT_ENCERRA']]

# Verificando o formato (tipo) das variáveis (colunas)
print(df_selecionado.dtypes)

# Convertendo as colunas para o formato datetime
df_selecionado['DT_DIAG'] = pd.to_datetime(df_selecionado['DT_DIAG'], format='%Y%m%d')
df_selecionado['DT_INIC_TR'] = pd.to_datetime(df_selecionado['DT_INIC_TR'], format='%Y%m%d')

# Verificando o novo formato das variáveis
print(df_selecionado.dtypes)

df_selecionado['DT_ENCERRA2'] = pd.to_datetime(
    df_selecionado['DT_ENCERRA'],
    format='%Y%m%d',
    errors='coerce'  # Trata erros e define valores inválidos como NaT
)

# Definindo como NaT as datas anteriores a 2000
df_selecionado['DT_ENCERRA2'] = df_selecionado['DT_ENCERRA2'].apply(
    lambda x: pd.NaT if x and x.year < 2000 else x
)

# Calculando a diferença em dias entre 'DT_DIAG' e 'DT_INIC_TR'
df_selecionado['TEMPO_DIAG_TRAT'] = (df_selecionado['DT_INIC_TR'] - df_selecionado['DT_DIAG']).dt.days
df_selecionado['TEMPO_TRAT_ENC'] = (df_selecionado['DT_ENCERRA2'] - df_selecionado['DT_INIC_TR']).dt.days
df_selecionado['TEMPO_DIAG_ENC'] = (df_selecionado['DT_ENCERRA2'] - df_selecionado['DT_DIAG']).dt.days

import numpy as np

# Condição para substituir 'TEMPO_TRAT_OBITO' por NaN quando for igual a 'TEMPO_DIAG_OBITO' e o valor for negativo
df_selecionado['TEMPO_TRAT_ENC'] = np.where(
    (df_selecionado['TEMPO_TRAT_ENC'] < 0),
    0,  # Substitui por NaN
    df_selecionado['TEMPO_TRAT_ENC']  # Caso contrário, mantém o valor original
)

# Convertendo as colunas para o formato datetime
df['DT_DIAG'] = pd.to_datetime(df['DT_DIAG'], format='%Y%m%d')
df['DT_INIC_TR'] = pd.to_datetime(df['DT_INIC_TR'], format='%Y%m%d')


# Criando uma nova coluna DT_ENCERRA2, convertendo a coluna DT_ENCERRA para datetime
df['DT_ENCERRA'] = pd.to_datetime(
    df['DT_ENCERRA'],
    format='%Y%m%d',
    errors='coerce'  # Trata erros e define valores inválidos como NaT
)

# Definindo como NaT as datas anteriores a 2000
df['DT_ENCERRA'] = df['DT_ENCERRA'].apply(
    lambda x: pd.NaT if x and x.year < 2000 else x
)

# Calculando a diferença em dias entre 'DT_DIAG' e 'DT_INIC_TR'
df['TEMPO_DIAG_TRAT'] = (df['DT_INIC_TR'] - df['DT_DIAG']).dt.days

# Calculando a diferença em dias entre 'DT_ENCERRA2' e 'DT_INIC_TR'
df['TEMPO_TRAT_ENC'] = (df['DT_ENCERRA'] - df['DT_INIC_TR']).dt.days

# Calculando a diferença em dias entre 'DT_ENCERRA2' e 'DT_DIAG'
df['TEMPO_DIAG_ENC'] = (df['DT_ENCERRA'] - df['DT_DIAG']).dt.days

# Condição para substituir 'TEMPO_TRAT_ENC' por NaN quando for igual a 'TEMPO_DIAG_ENC' e o valor for negativo
df['TEMPO_TRAT_ENC'] = np.where(
    (df['TEMPO_TRAT_ENC'] < 0),
    0,  # Substitui por NaN
    df['TEMPO_TRAT_ENC']  # Caso contrário, mantém o valor original
)

```

```python
# Get current columns after dropping some in previous steps
current_cols = df.columns

# Filter colunas to only include columns that still exist in df
cols_to_process = [col for col in current_cols if col in current_cols and col not in ['NU_ANO','pop']]

for col in cols_to_process:
    df[col] = df[col].replace(9, np.nan)


for col in cols_to_process:
    df[col] = df[col].replace(99, np.nan)

for col in cols_to_process:
    # Check if the column is of object type before replacing empty strings
    if df[col].dtype == 'object':
        df[col] = df[col].replace('', np.nan)

# Verificar a presença de valores nulos
print(df.isnull().sum())

# Print the list of columns being processed (optional, for debugging)
# print("Columns being processed:", cols_to_process)
```
